```python
import pandas as pd

# Loading dataset
df = pd.read_csv("owid-covid-data.csv")

# Inspecting basic structure
print(df.columns)  # View column names
print(df.head())   # Preview first five rows
print(df.info())   # View data types
print(df.isnull().sum())  # Checking missing values

# Converting date column to datetime format
df["Date_reported"] = pd.to_datetime(df["Date_reported"])

# Selecting countries of interest
countries = ["Kenya", "USA", "India"]
df_filtered = df[df["Country"].isin(countries)]

# Handling missing values (e.g., fill missing cases with interpolation)
df_filtered.fillna(method="ffill", inplace=True)

print(df_filtered.head())  # Check cleaned data

import matplotlib.pyplot as plt
# Ploting total cases over time
for country in countries:
    country_df = df_filtered[df_filtered["Country"] == country]
    plt.plot(country_df["Date_reported"], country_df["Cumulative_cases"], label=country)

plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.title("COVID-19 Total Cases Over Time")
plt.legend()
plt.show()

```
